## Gráfico de Linhas
Vamos criar um gráfico de linhas para analisar a evolução do preço das ações das empresas Tesla e Apple.

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ticker =  yf.Ticker('TSLA')
df = ticker.history(interval='1d', start= '2020-01-01')
```

```python
df.head()
```

```python
ticker =  yf.Ticker('AAPL')
df2 = ticker.history(interval='1d', start= '2020-01-01')
```

```python
df['Close AAPL'] = df2['Close']
df.rename(columns={'Close': 'Close TSLA'}, inplace=True)
df.head()
```

```python
plt.rcParams['figure.figsize'] = [12, 8]
sns.set_style('darkgrid')
df[['Close TSLA', 'Close AAPL']].plot()
plt.title('Açoes da tesla e apple')
plt.xlabel('Tempo')
plt.ylabel('Valor')
plt.show()
```

Vamos agora utilizar o dataset LifeExpectancy para criar um gráfico de barras.

```python
df3 = pd.read_csv('lifeExpectancy.csv')
df3.head()
```

Nosso gráfico deve permitir comparar a expectativa de vida de homens e mulheres de países africanos.

```python
df3.plot(x='Countries', kind='bar', stacked=False)
plt.title('Expectativa de de vida H x M na Africa')
plt.xlabel('Paise')
plt.ylabel('Expectativa de vida')
plt.show()
```

Para uma melhor visualização, podemos dividir o dataset. A seguir vamos tomar os 15 primeiros países do dataset.

```python
df4 =df3.head(10)
df4.plot(x='Countries', kind='bar', stacked=False)
plt.title('Expectativa de de vida H x M na Africa')
plt.xlabel('Paise')
plt.ylabel('Expectativa de vida')
plt.show()
```

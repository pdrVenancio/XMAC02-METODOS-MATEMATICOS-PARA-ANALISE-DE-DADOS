Questão 1 \
Utilizando o dataset 'tips' crie um gráfico de barras mostrando o faturamento total do restaurante por dia, exibindo o total gasto por homens e mulheres. \
Dica: use crosstab.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
# Criar uma tabela cruzada (crosstab) com os totais gastos por dia e gênero
faturamento = pd.crosstab(df['day'], df['sex'], values=df['total_bill'], aggfunc='sum')

# Plotar o gráfico de barras
faturamento.plot(kind='bar', stacked=True)

# Adicionar título e rótulos
plt.title('Faturamento Total por Dia e Gênero')
plt.xlabel('Dia da Semana')
plt.ylabel('Faturamento Total')
plt.legend(title='Gênero')

# Exibir o gráfico
plt.show()
```

Questão 2 \
Elabre um gráfico de pizza que mostre a divisão do faturamento entre homens e mulheres.

```python
# Calcular o total faturado por homens e mulheres
faturamento_por_genero = df.groupby('sex', observed=False)['total_bill'].sum()

# Criar o gráfico de pizza
plt.pie(faturamento_por_genero, labels=faturamento_por_genero.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])

# Adicionar título
plt.title('Divisão do Faturamento entre Homens e Mulheres')

# Exibir o gráfico
plt.axis('equal')  # Garantir que o gráfico seja um círculo
plt.show()
```

Questão 3 \
Elabore um gráfico de dispersão que mostre a relação entre o número de pessoas por mesa e o valor total da conta.

```python
# Criar o gráfico de dispersão
plt.scatter(df['size'], df['total_bill'], alpha=0.7, c='blue')

# Adicionar título e rótulos aos eixos
plt.title('Relação entre o Número de Pessoas por Mesa e o Valor Total da Conta')
plt.xlabel('Número de Pessoas por Mesa')
plt.ylabel('Valor Total da Conta')

# Exibir o gráfico
plt.show()
```

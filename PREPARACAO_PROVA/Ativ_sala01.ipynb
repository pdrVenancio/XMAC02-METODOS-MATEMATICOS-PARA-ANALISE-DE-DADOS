Questão 1 \
Suponha que uma central telefônica recebeu 270 chamadas num período de 3 horas, ou seja, 1,5 chamadas por minuto. Deseja-se calcular a probabilidade de que nos próximos 3 minutos sejam recebidas 0, 1, 2, ..., 8 chamadas. Mostre essas probabilidades plotando um gráfico de barras (barplot).

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom, poisson, norm
```

```python
eixo_x = np.arange(0,9)
eixo_y = poisson.pmf(eixo_x, 4.5)
sns.barplot(x=eixo_x, y=eixo_y)
plt.show()
```

Questão 6 \
Uma linha de produção está sendo analisada para efeito de controle de qualidade das peças produzidas. Tendo em vista o alto padrão requerido, a produção é interrompida para regulagem toda vez que uma peça defeituosa é observada. Se 0,01 é a probabilidade da peça ser defeituosa, estude o comportamento da variável Q, quantidades de peças boas produzidas antes da primeira defeituosa. Para tanto, simule a geração de peças e conte quantas são produzidas até que a primeira defeituosa seja encontrada. Repita o processo 1.000 vezes e calcule a média.
A média vai representar o número de peças boas são produzidas até que uma peça defeituosa seja produzida. 

```python
lista_cont = []
for i in range(1000):
    cont = 0
    while True:
        if np.random.binomial(1, 0.01) == 0:
            cont = cont+1
        else:
            break
    lista_cont.append(cont)
    #print(cont)
vet = np.array(lista_cont)
vet.mean()
```

Questão 8 \
Uma fábrica engarrafa bebidas contendo 600 ml e desvio padrão de 12 ml. Um lote de 1000 garrafas foi produzido. Sem utilizar a linguagem Python, informe: \
A) A quantidade de garrafas contendo mais de 612 ml. \
B) A quantidade de garrafas contendo entre 588 e 624 ml. \
Explique como chegou nesses resultados. Lembre-se que não é permitido obter a resposta usando funções do Python.

```python
# Dados do problema
media = 600  # média em ml
desvio_padrao = 12  # desvio padrão em ml
n_garrafas = 1000  # número de garrafas no lote

# A) Quantidade de garrafas contendo mais de 612 ml
limite_superior = 612

# Calcula o valor Z para 612 ml
z_612 = (limite_superior - media) / desvio_padrao

# Calcula a probabilidade de uma garrafa ter mais de 612 ml
p_mais_612 = 1 - norm.cdf(z_612)

# Calcula a quantidade de garrafas com mais de 612 ml
qtd_mais_612 = p_mais_612 * n_garrafas

# B) Quantidade de garrafas contendo entre 588 e 624 ml
limite_inferior = 588
limite_superior_2 = 624

# Calcula o valor Z para 588 ml e 624 ml
z_588 = (limite_inferior - media) / desvio_padrao
z_624 = (limite_superior_2 - media) / desvio_padrao

# Calcula a probabilidade de uma garrafa estar entre 588 ml e 624 ml
p_entre_588_624 = norm.cdf(z_624) - norm.cdf(z_588)

# Calcula a quantidade de garrafas entre 588 ml e 624 ml
qtd_entre_588_624 = p_entre_588_624 * n_garrafas

# Resultados
print(f"A) Quantidade de garrafas contendo mais de 612 ml: {round(qtd_mais_612)}")
print(f"B) Quantidade de garrafas contendo entre 588 ml e 624 ml: {round(qtd_entre_588_624)}")
```

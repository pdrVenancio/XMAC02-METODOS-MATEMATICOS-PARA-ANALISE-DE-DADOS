```python
# import libraries
import pandas as pd
import numpy as np

# data visualization
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

### Questão 1
Gere um gráfico de barras que mostre a quantidade de homens e mulheres com risco alto e com risco baixo de ataque cardíaco. 

```python
# Criar uma tabela cruzada (crosstab) com os totais gastos por dia e gênero
grafico = pd.crosstab(df['output'], df['sex'], values=df['sex'], aggfunc='count')

# Plotar o gráfico de barras
grafico.plot(kind='bar', stacked=False)

# Adicionar título e rótulos
plt.title('Heart Attack Risck vs Sex')
plt.xlabel('Risck')
plt.ylabel('Aumont')
plt.legend(title='sex')

# Exibir o gráfico
plt.show()
```

### Questão 2
Faça dois gráficos de pizza lado a lado que mostre como homens e mulheres se dividem percentualmente em termos de risco cardíaco.

```python
df = pd.read_csv('heart.csv')

m = df[df['sex'] == 'Male']
w = df[df['sex'] == 'Female']

wva = w['output'].value_counts()
mva = m['output'].value_counts()

plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1)
sns.set_style('darkgrid')
plt.pie(wva, labels=wva.index, autopct='%1.1f%%', startangle=150)
plt.title('Distribuição de resultados cardíacos em mulheres')
plt.axis('equal')


plt.subplot(1, 2, 2)
sns.set_style('darkgrid')
plt.pie(mva, labels=mva.index, autopct='%1.1f%%', startangle=300)
plt.title('Distribuição de resultados cardíacos em homens')
plt.axis('equal')

plt.tight_layout()
plt.show()
```

### Questão 3 
Faça um gráfico de pizza que mostre a divisão de risco cardíaco (High Risk) entre homens e mulheres, de acordo com os dados do dataset. Note que o dataset tem mais dados de homens, o que impacta o resultado.

### Questão 4
Faça dois gráficos de barras lado a lado que mostrem as distribuições de idade de homens e mulheres do dataset.

```python
masculino = df[df['sex'] == 'Male']['age']
feminino = df[df['sex'] == 'Female']['age']

fig, axs = plt.subplots(1, 2, figsize=(12, 6), sharey=False)

sns.histplot(masculino, bins=8, ax=axs[0], color='blue', kde=False)
axs[0].set_title('Age distribution (Men)')

sns.histplot(feminino, bins=8, ax=axs[1], color='green', kde=False)
axs[1].set_title('ge distribution (Men)')

plt.tight_layout()
plt.show()
```

Considerando o colesterol (chol) dos pacientes do dataset, retire outliers com valores acima de 500. Em seguida, mostre um gráfico de dispersão relacionando a idade e o valor do colesterol dos paciente do dataset.

### Questão 6
A coluna cp significa "chest pain" (dor no peito). Existem 4 tipos de cp (0 - 3). Deseja-se saber se há relação entre a pressão arterial do paciente (trtbps) e o tipo de dor no peito apresentado. Para tanto, faça um gráfico de barras que mostre a pressão arterial média dos pacientes de cada categoria de dor (0 - 3), dividido por sexo.

```python

df = pd.read_csv('heart.csv')
df2 = pd.crosstab(df['sex'], df['cp'], values=df['trtbps'], aggfunc='mean')


plt.rcParams['figure.figsize'] = [10,6]
sns.set_style('darkgrid')
df2.plot(kind='bar', stacked=False)
plt.title('Migrante desaparecidos')
plt.ylabel('Quantidades de desaparecidos')
plt.xlabel('Regiao')
plt.show()
```
